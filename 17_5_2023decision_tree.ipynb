```python
# for data preprocecing
import pandas as pd
import numpy as np
```

```python
# import the dataset

df = pd.read_csv(r"/content/salaries.csv")
df
```

```python
# DIVIDE A FEATURE FROM DF
featur = df.drop('salary_more_then_100k',axis = 1)   # feature
featur
```

```python
# target
target = df['salary_more_then_100k']
target
```

```python
# labelencoder import

from sklearn.preprocessing import LabelEncoder
```

```python
# encode for all feature column

cmp = LabelEncoder()
job = LabelEncoder()
deg =LabelEncoder()
```

```python
featur['cmp_enc'] = cmp.fit_transform(featur['company'])
featur['job_enc'] = job.fit_transform(featur['job'])
featur['deg_enc'] = deg.fit_transform(featur['degree'])
```

```python
featur
```

```python
# create a final feature

final_feat = featur.drop(['company','job','degree'],axis=1)
final_feat
```

```python
target
```

```python
# Now We are Create a decision tree
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(final_feat,target)
```

```python
model.score(final_feat,target)
```

```python
model.predict([[2,0,0]]) # suppose we are predict
```
